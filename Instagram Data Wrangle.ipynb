```python
import pandas as pd
import json
import urllib
from IPython.display import display
```

```python
# Import the insta Links of labelled data
allurl = pd.read_csv(r'C:\Users\acer\Desktop\Interview\Winkl\Task 2\Data Created\givendata.csv')
allurl.head()
```

```python
# Attributes we want to extract
attributes = ['user','bio']
# Storing dataframe
DATABASE = pd.DataFrame([["",""]], columns = attributes)
# GET request string
getreq = '/?__a=1'

# Searching through each link for their bio
for x in allurl['Instagram']:
    url = x+getreq
    response = urllib.request.urlopen(url)
    data = json.loads(response.read())
    bio = data["graphql"]["user"]["biography"]
    name = data["graphql"]["user"]["username"]
    row = {'user':name, 'bio':bio}
    DATABASE = DATABASE.append(row, ignore_index = True)
```

```python
display(DATABASE)
# Save data to csv file
DATABASE.to_csv(r'C:\Users\acer\Desktop\Interview\Winkl\Task 2\Data Created/fashion.csv')
```
